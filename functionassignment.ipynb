```python
#6. Write a Python function that accepts a string and counts the number of upper and lower case letters.
	#Sample String : 'The quick Brow Fox'
def count_no(s):
    upper = 0
    lower = 0

    for i in s:
        if i.isupper():
            upper += 1
        elif i.islower():
            lower += 1

    return upper, lower

# Example
x = "The quick Brow Fox"
u,l=count_no(x)
print("Uppercase letters:", u)
print("Lowercase letters:", l)
```

```python
#7. Write a Python function that takes a list and returns a new list with distinct elements from the first list.
#Sample List : [1,2,3,3,3,3,4,5]

def distinct_list(lst):
    distinct_list = []

    for item in lst:
        if item not in distinct_list:
            distinct_list.append(item)

    return distinct_list

# Example
data = [1,2,3,3,3,3,4,5]
x=distinct_list(data)

print("Distinct elements:", x)

#method2
def distinct_lst(lst):
    return list(set(lst))        #does not preserve order

data=[1,2,3,3,3,3,4,5]
x=distinct_lst(data)
print(x)
```

```python
#8. Write a Python function that checks whether a passed string is a palindrome or not.
def palin(text):
  if len(text)<=1:
    print("pelindrome")
  else:
    if text[0]==text[-1]:
      palin(text[1:-1])
    else:
      print("no palindrome")

palin("madam")

method2
def palin(text):
  return text==text[::-1]

palin('madam')
palin('amit')

```

```python
#9. Write a Python program that accepts a hyphen-separated sequence of words as input and
#prints the words in a hyphen-separated sequence after sorting them alphabetically.

#	Sample Items : green-red-yellow-black-white

def sort_list(sequence):
    words = sequence.split('-')   # Step 1: split into list
    words.sort()                  # Step 2: sort alphabetically
    return '-'.join(words)       # Step 3: join back with '-'

# Sample input
x = "green-red-yellow-black-white"

# Function call
result = sort_list(x)

print("Sorted sequence:", result)

```

```python
#22. Write a Python program to check whether a given string contains a capital letter, a lower case letter, a number and a minimum length using lambda.

check_string = lambda s: (
    any(map(lambda x: x.isupper(), s)) and
    any(map(lambda x: x.islower(), s)) and
    any(map(lambda x: x.isdigit(), s)) and
    len(s) >= 8
)

text = "Amit12345"

if check_string(text):
    print("Valid string")
else:
    print("Invalid string")
```

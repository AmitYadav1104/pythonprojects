```python
#5. You are given a list of integers. You are asked to make a list by running through elements of the list by adding all elements greater and itself.

#i.e. Say given list is [2,4,6,10,1] resultant list will be [22,20,10,23].

list1 = [2, 4, 6, 10, 1]

result = []

for i in range(len(list1)):
    total = list1[i]   # include itself

    for j in range(len(list1)):
        if list1[j] > list1[i]:
            total += list1[j]

    result.append(total)

print(result)
```

```python
#6. Find list of common unique items from two list. and show in increasing order


	#num1 = [23,45,67,78,89,34]
	#num2 = [34,89,55,56,39,67]

num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]

result = []

for i in num1:
    if i in num2 and i not in result:
        result.append(i)

result.sort()
print(result)
```

```python
#7. Sort a list of alphanumeric strings based on product value of numeric character in it.
# If in any string there is no numeric character take it's product value as 1.


data = ['1ac21', '23fg', '456', '098d', '1', 'kls']

def product_num(s):
    prod = 1
    has_digit = False

    for ch in s:
        if ch.isdigit():
            prod *= int(ch)
            has_digit = True

    return prod if has_digit else 1

result = sorted(data, key=product_num, reverse=True)

print(result)
```

```python
#8. Write a program that can find the max number of each row of a matri

	#[[1,2,3],[4,5,6],[7,8,9]]

def max_num(data):
    result = []
    for i in data:
        result.append(max(i))
    return result

data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

print(max_num(data))
```

```python
#12. Intersection of two lists. Intersection of two list means we need to take all those elements which are common to both of the initial lists and store them into another list. Only use using list-comprehension.

lst1 = [15, 9, 10, 56, 23, 78, 5, 4, 9]
lst2 = [9, 4, 5, 36, 47, 26, 10, 45, 87]

result = [i for i in lst1 if i in lst2]

print("Intersection:", result)
```
